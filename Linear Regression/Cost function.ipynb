# Cost Function

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error as mse
```

# Creating Sample Data

```python
exp=[1.2,1.5,1.9,2.2,2.4,2.5,2.8,3.1,3.3,3.7,4.2,4.4]
sal=[1.7,2.4,2.3,3.1,3.7,4.2,4.4,6.1,5.4,5.7,6.4,6.2]

data=pd.DataFrame({
    'sal':sal,
    'exp':exp
})
data.head()
```

```python
# Plotting the curve

plt.scatter(data.exp,data.sal,color='red',label='data points')
plt.xlim(1,4.5)
plt.ylim(1,7)
plt.xlabel('exp')
plt.ylabel('sal')
plt.legend()
```

# Starting the lines using small values of parameters

```python
# Making lines for diff values of beta 0.1,0.8,1.5
beta=0.8
#Keeping intercept constant
b=1.1
# To store predicted points
line1=[]

# generating predictions for every data point
for i in range(len(data)):
    line1.append(data.exp[i]*beta+b)
    
# Plotting the line
plt.scatter(data.exp,data.sal,color='red')
plt.plot(data.exp,line1,color='black',label='line')
plt.xlim(1,4.5)
plt.ylim(1,7)
plt.xlabel('exp')
plt.ylabel('sal')
plt.legend()
MSE=mse(data.exp,line1)
plt.title('Beta value '+str(beta)+"with mse"+str(MSE))
MSE=mse(data.exp,line1)
```

# Computing Cost over a range of values of Beta

```python
# function to calculate error

def Error(Beta,data):
    # b is constant
    b=1.1
    
    sal=[]
    exp=data.exp
    
    # Loop to calculate predict salary variables
    for i in range(len(data.exp)):
        tmp=data.exp[i]*Beta+b
        sal.append(tmp)
    MSE=mse(exp,sal)
    return MSE
```

```python
# range of slopes from 0 to 1.5 with increment of 0.01
slope=[i/100 for i in range(0,150)]
Cost=[]
for i in slope:
    cost=Error( Beta = i,data = data)
    Cost.append(cost)
```

```python
# Arranging in DataFrame

cost_table=pd.DataFrame({
    'Beta':slope,
    'Cost':Cost
})
cost_table.head()
```

# visualizing cost wrt Beta

```python
# Plotting the cost values corresponding to every value of beta
plt.plot(cost_table.Beta,cost_table.Cost,color='blue',label='Cost Function Curve')
plt.xlabel('Value of Beta')
plt.ylabel('Cost')
plt.legend()
```
