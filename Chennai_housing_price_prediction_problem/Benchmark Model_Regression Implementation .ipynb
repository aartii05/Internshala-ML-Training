# Shufflingg and Creating Train and Test Set.For that I will shuffle entire dataset divide into 4 parts.Out of these 4 parts I will use 3 parts for my train dataset and 1 part for test dataset

```python
from sklearn.utils import shuffle

#Shuffling the dataset
df=shuffle(df,random_state=42)

#Creating 4 divisions
div=int(df.shape[0]/4)

# 3 parts to Train set and 1 part to test set
train=df.loc[:3*div+1,:]
test=df.loc[3*div+1:]
```

```python
train.head()
```

```python
test.head()
```

# Simple Mean (mean of Item_Outlet_Sales)

```python
# Storing simple mean in a new column in the test set as "simple_mean"
test['simple_mean']=train['Item_Outlet_Sales'].mean()
```

```python
test.head()
```

# Calculating mean absolute error

```python
from sklearn.metrics import mean_absolute_error as MAE

simple_mean_error=MAE(test['Item_Outlet_Sales'],test['simple_mean'])
simple_mean_error
```
