## Dataframe and its operation

### A DataFrame is similar to Excel workbook tabular datasheet. 

### |Operation|Command|
### |Dimensions of dataframe"row x cols"|DataFrame.shape()|
### |Access Top/Bottom<n> records|DataFrame.head(n)/DataFrame.tail(n)
### |Access all cols names|DataFrame.cols|
### |Access data of one col|DataFrame['colname']|
### |Access data of multiple cols|DataFrame[['col1','col2',.....]]
